Don't have twitter API so I used apify to scrape Andrew Tate quotes off of twitter. Here I'm putting everything in a nice dataframe.

```python
# create an empty list to store the data
rows = []

# iterate through the json data and extract the desired information
for item in tqdm(data):
    full_text = item["full_text"]
    if full_text[:2]=='RT' or full_text[:5]=='https':
        continue
    name = item["user"]["name"]
    date = item["created_at"]
    url = item["url"]
    # add the extracted data as a row to the list
    rows.append([name, full_text, date, url])

for tenet in tenets:
  rows.append(['Andrew Tate',tenet,'NA','NA'])

df = pd.DataFrame(rows, columns=["name", "full_text", "date", "url"])
df.to_csv(DIR+"tate.csv")
```

Goodreads is super easy to scrape. Managed to get a lot of Confucius quotes off of there.

```python
confucius = []
for i in tqdm(range(17)):
    url = "https://www.goodreads.com/author/quotes/15321.Confucius?page=" + f'{i+1}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    quotes = soup.find_all('div',{'class':'quoteText'})
    for quote in quotes:
        text = re.sub(r'\n', '', str(quote)) # remove newline characters
        # text = re.sub(r'Confucious', '', text) # remove the word "Confucious"
        # text = re.sub(r'Confucius', '', text) # remove the word "Confucious"
        # text = re.sub(r'\s+', ' ', text) # remove extra spaces
        # text = re.sub(r'-', '',text)
        # text = re.sub(r'―', '',text)
        text = re.search(r'<div class="quoteText">(.+?)<br/>',text)
        text = str(text.group(1))
        text = text.strip() # remove whitespace from the beginning and end
        text = text.strip('“').strip('”')
        # text = re.findall(r'".+"', text)
        confucius.append(['Confucius',text,'NA',url])

df1 = pd.DataFrame(confucius, columns=["name", "full_text", "date", "url"])
```

```python
def print_optional_kwargs(func):
    import inspect
    # Get the list of arguments of the input function
    args = inspect.getfullargspec(func).args
    # Get the list of default values of the input function
    defaults = inspect.getfullargspec(func).defaults
    # Check if there are any default values
    if defaults:
        # Iterate through the list of arguments and default values
        for i, arg in enumerate(args[-len(defaults):]):
            print(f"{arg} (optional)")
    else:
        print("No optional keyword arguments.")
```

```python
quotes_and_logprobs = {}
tate_quotes = list(pd.read_csv('tate.csv').loc[:,'full_text'])
```

```python
from tqdm import tqdm
import time
from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)  # for exponential backoff
```

```python
@retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(6))
def completion_with_backoff(**kwargs):
    return openai.Completion.create(**kwargs)

time.sleep(120)
quotes_and_logprobs = []
count = 0
for quote in tqdm(tate_quotes):

    time.sleep(8)
    result = completion_with_backoff(
      model="code-davinci-002",
      prompt=f"This sentence sounds like Confucius: {quote}",
      max_tokens=0,
      temperature=0,
      logprobs=0,
      echo=True
    )

    probability = sum(result.choices[0].logprobs.token_logprobs[1:])
    all_logprobs = result.choices[0].logprobs.token_logprobs
    quotes_and_logprobs.append([quote, probability, all_logprobs])

# df = pd.DataFrame(quotes_and_logprobs, columns=["quote", "probability", "logprobs"])
# df.to_csv('tate_probs.csv')
# df.sort_values(by='probability',ascending=False).iloc[20:60,]
```
